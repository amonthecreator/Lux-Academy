### Question 1
#### Find the number of times the wind speed was exactly 4 km/hr.

```python
print("The number of data points having wind speed of 4 km/h is: ",len(df[df['Wind Speed_km/h'] == 4]))
```

### Question 3
#### Rename the column "Weather" to "Weather_Condition."

```python
df = df.rename(columns = {'Weather':'Weather_Condition'})
df
```

### Question 4
#### What is the mean visibility of the dataset?

```python
print("The mean value of visibility in km is : ",df['Visibility_km'].mean())
```

### Question 5
#### Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.

```python
wind_speed = df[df['Wind Speed_km/h'] > 24]
both = wind_speed[wind_speed['Visibility_km'] == 25]

print("The number of records is: ",len(both.value_counts()))
```

### Question 6
#### What is the mean value of each column for each weather condition?

```python
means_by_weather = df.groupby('Weather_Condition').mean()
means_by_weather
```

### Question 7
#### Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.

```python
filtered_weather = df[(df['Weather_Condition'] == 'Clear') &
    ((df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40))
]

filtered_weather
```

### Question 8
#### Find the number of weather conditions that include snow.
